```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x1=[1,5,7]
y1=[2,4,9]
x2=[5,6,7]
y2=[7,8,9]
plt.plot(x1,y1,label="First")
plt.plot(x2,y2,label="Second")
plt.legend()
plt.show()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')

x1=[1,5,7]
y1=[2,4,9]
x2=[5,6,7]
y2=[7,8,9]
plt.plot(x1,y1,label="First",linewidth=10)
plt.plot(x2,y2,label="Second",linewidth=5)
plt.legend()
plt.show()
```
